```python
import numpy as np
# read inspired_SVT.txt
with open('inspired_SVT.txt', 'r') as f:
    inspired_SVT = f.read()
```

```python
times = inspired_SVT.split("\n")
```

```python
for i in times:
    if i.startswith("SVD"):
        print(i)
```

```python
data_construction_time = 0
for i in times[:-1]:
    data_construction_time += (float(i.split(":")[-1]))
```

```python
float(times[-1].split(":")[-1]) - data_construction_time
```

```python
import pandas as pd

data = []
lines = inspired_SVT.split('\n')
for i in range(0, len(lines), 2):
    elapsed_time = float(lines[i].split(': ')[1])
    rmse = float(lines[i+1].split(': ')[1])
    rmse_info_2 = lines[i+1].split(': ')[0]
    #print(rmse_info_2)
    rank = int(rmse_info_2.split(' with ')[0].split()[-1])
    l = int(rmse_info_2.split('l=')[1].split(',')[0])
    p = int(rmse_info_2.split('p=')[1].split(':')[0])
    #rmse = float(rmse_info.split(': ')[1])
    data.append({'Method': rmse_info_2.split( )[0],'Elapsed Time': elapsed_time, 'RMSE': rmse, 'Rank': rank, 'l': l, 'p': p})

df = pd.DataFrame(data)
print(df)
```
